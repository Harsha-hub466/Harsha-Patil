```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Iris.csv')
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
plt.title('scatter plot')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```

```python
import pandas as pd
species_counts = df['Species'].value_counts()
plt.bar(species_counts.index, species_counts.values, color='lightblue')
plt.title("Number of Flowers by Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.grid(axis='y')
```

```python
plt.hist(df['PetalLengthCm'], bins=10, color='lightgreen', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
species = df['Species'].unique()
data = [df[df['Species'] == s]['PetalLengthCm'] for s in species]
plt.boxplot(data, labels=species)
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```
